```python
# dropping so2 features 
data = data.drop(columns=['SO2', 'NOX'])
data.head()
data.isnull()
```

```python
# data shape
data.shape
```

```python
# drop isnull values
data = data.dropna()
data.shape
```

```python
# separating numerical features and categorical features
num_cols = data.select_dtypes(include=[np.number]).columns
cat_cols = data.select_dtypes('object').columns

print(num_cols)
print(cat_cols)

```

```python
# splitting data into train and test
from sklearn.model_selection import train_test_split

train , test = train_test_split(data, test_size=0.2, random_state=42)

train_data , val_data = train_test_split(train, test_size=0.2, random_state=42)



```

```python
X_train = train_data.drop(columns=['Temperature'])
y_train = train_data['Temperature']

X_val = val_data.drop(columns=['Temperature'])
y_val = val_data['Temperature']

num_cols = X_train.select_dtypes(include=[np.number]).columns
cat_cols = X_train.select_dtypes('object').columns


print(num_cols)
print(cat_cols)

```

```python
# imputing missing values with simple imputer
from sklearn.impute import SimpleImputer

num_imputer = SimpleImputer(strategy='mean')
cat_imputer = SimpleImputer(strategy='most_frequent')

X_train[num_cols] = num_imputer.fit_transform(X_train[num_cols])
X_val[num_cols] = num_imputer.transform(X_val[num_cols])




```

```python
# scaling numerical features and encoding categorical features
from sklearn.preprocessing import StandardScaler, OrdinalEncoder

scaler = StandardScaler()
encoder = OrdinalEncoder()
X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_val[num_cols] = scaler.transform(X_val[num_cols])

X_train[cat_cols] = encoder.fit_transform(X_train[cat_cols])
X_val[cat_cols] = encoder.transform(X_val[cat_cols])

```

```python
#building model
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)


```

```python
#evaluating model
from sklearn.metrics import root_mean_squared_error, r2_score

y_pred = model.predict(X_val)
rmse = root_mean_squared_error(y_val, y_pred)
print(f"Root Mean Squared Error: {rmse}")

```
